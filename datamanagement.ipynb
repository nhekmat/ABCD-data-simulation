```python
# import arrow
import pandas as pd
```

```python
abcd = pd.read_csv('C:\\Users\\kbaacke\\ABCD_6.0\\HeatProject_01\\dataset.csv')
abcd
```

```python
# Filter to only include sessions with DERS
abcd_f = abcd.loc[~pd.isna(abcd['sdev_y_ders_001'])]
abcd_f
```

```python
# Make index DF

## Make a list of columns where all values in filtered data are NA
index_cols = ['participant_id','session_id']
for c in list(abcd_f.columns):
    if len(abcd_f[~pd.isna(abcd_f[c])])==0:
        print(f'Column {c} is not avaialable in this subset; creating indexing column and dropping from subset data.')
        abcd_f = abcd_f.drop(c, axis=1)
        index_cols.append(c)
        
index_cols
```

```python
abcd_f['session_id'].unique()
```

```python
index_df = abcd[index_cols]
index_df['session_id_alt'] = ''
index_df.loc[index_df['session_id']=='ses-01A','session_id_alt'] = 'ses-S01'
index_df = index_df.loc[index_df['session_id_alt']!='']
index_df = index_df.drop('session_id', axis=1)
index_df.rename(columns={
    'session_id_alt':'session_id'
}, inplace=True)
index_df
```

```python
# Merge in indexed columns

new_df = pd.merge(abcd_f, index_df, how='left', on=['participant_id','session_id'])
new_df
```
